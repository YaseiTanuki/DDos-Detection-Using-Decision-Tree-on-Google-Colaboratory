# Get Started

### Mount drive
Take dataset from my google drive. The path can be change.

```python
from google.colab import drive

# Mount drive
drive.mount('/content/drive')

# Path to dataset on drive
path = "/content/drive/MyDrive/Dataset/Syn0311.csv"
```

### Use Pandas
Display all columns (features) and treat infinity value as NaN to easily remove.

```python
import pandas as pd

# Pandas option
pd.set_option('display.max_columns', None)
pd.set_option('mode.use_inf_as_na', True)
```

### Read data
Colab has limit resources, reading 1 000 000 rows is a good deal.

```python
# Read data (SimillarHTTP column has mixed type)
df = pd.read_csv(path, dtype={'SimillarHTTP': object}, nrows=1000000)
```

# Clean Data

Remove whitespace from columns name

```python
# Remove space from columns name
df.columns = df.columns.str.replace(' ', '')

# Print top 5 samples
df.head(5)
```

**Drop irrelevant features**

```python
df.drop(columns=['Unnamed:0', 'FlowID', 'SourceIP', 'DestinationIP', 'Timestamp', 'SimillarHTTP', 'SourcePort', 'DestinationPort'], inplace=True)
df.head(5)
```

Remove NaN value (infinity value now also treat as NaN)

```python
df.dropna(inplace=True)
```

# Cleaned Dataset Infomation

```python
df.info()
```

# Preprocessing

Plit dataset into Features and Label

```python
# using X as input (without label)
X = df.drop(columns=['Label'])
# using y as output (label)
y = df['Label']

target_names = ['Syn', 'BENIGN']
```

Number of features and samples

```python
# display number of samples and features
print(f"X: {X.shape}")
X.head()
```

## Resample Data

import matplolib for drawing chart, and seaborn for color template.

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

Check dataset balance

```python
plt.figure(figsize=(8,8))
plt.pie(y.value_counts(), labels=target_names, autopct='%1.2f%%', colors=sns.color_palette('Set2'), textprops={'fontsize':11})
plt.title(
    label="Dataset Labels Before Resample",
    fontdict={"fontsize":16},
)

plt.show()
```

Dataset is unbalance with may harm evaluation result. The way to deal with this is resample dataset. SMOTE is use for up-resampling.

```python
# SMOTE
from imblearn.over_sampling import SMOTE

sm = SMOTE(random_state=42)
X_res, y_res = sm.fit_resample(X, y)
```

Check dataset balance again

```python
plt.figure(figsize=(8,8))
plt.pie(y_res.value_counts(), labels=target_names, autopct='%1.2f%%', colors=sns.color_palette('Set2'), textprops={'fontsize':11})
plt.title(
    label="Dataset Labels after resampling",
    fontdict={"fontsize":16},
)

plt.show()
```

## Features Selection

Not all 79 features will be used, feature selection will choose features that strongly affect model result.


Using SelectFromModel for features selection, SelectFromModel require an estimator as input so ExtraTreesClassifier is choosen

```python
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.feature_selection import SelectFromModel
```

```python
estimator = ExtraTreesClassifier(n_estimators=50)
estimator.fit(X_res, y_res)
```

```python
# max_features is maximum number of feature that will be choosen
# The number of choosen feature can be lower than max_feature
max_feature = 2
selector = SelectFromModel(estimator, max_features=max_feature)
```

```python
feature_name = selector.get_feature_names_out(X.columns)
print(f'Feature {feature_name} is choosen')
```

```python
plt.rcParams["font.family"] = "monospace"
plt.title("Feature importance")
feat_importances = pd.Series(estimator.feature_importances_, index=X.columns)
plt.ylabel('Feature name')
feat_importances.nlargest(5).plot(kind='barh', color=['#9CCFE7', '#FDD0B1'])
```

```python
X_res = X_res[feature_name]
X_res.head(5)
```

# Training

Split data to training set and testing set

```python
from sklearn.model_selection import train_test_split

# Spliting dataset into training and testing dataset.
X_train, X_test, y_train, y_test = train_test_split(X_res, y_res, test_size=0.4)
```

Choose model and train

```python
from sklearn.tree import DecisionTreeClassifier

# Choose model
DTmodel = DecisionTreeClassifier()

# Train
DTmodel.fit(X_train, y_train)
```

```python
from sklearn.ensemble import RandomForestClassifier

# Choose model
RFmodel = RandomForestClassifier()

# Train
RFmodel.fit(X_train, y_train)
```

LinearSVC require Scaler to work efficient

```python
from sklearn.svm import LinearSVC
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler

# Choose model
LinearSVCmodel = make_pipeline(StandardScaler(), LinearSVC(dual=True, random_state=0, tol=1e-5))

# Train
LinearSVCmodel.fit(X_train, y_train)
```

# Evaluation

Make prediction

```python
# Decision Tree
DTprediction = DTmodel.predict(X_test)
```

```python
# LinearSVC
SVCprediction = LinearSVCmodel.predict(X_test)
```

```python
# Random Forest
RFprediction = RFmodel.predict(X_test)
```

Draw confusion matrix

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay
# matplotlib.pyplot is imported in previous section, if not, import it here
# import matplotlib.pyplot as plt
```

```python
# Confusion matrix
DTcm = confusion_matrix(y_test, DTprediction, labels=target_names)
DTcm_display = ConfusionMatrixDisplay(confusion_matrix = DTcm, display_labels=target_names)
DTcm_display.plot(cmap="RdPu")
plt.title(f"Decision Tree CM using {max_feature} features")

plt.show()
```

```python
# Confusion matrix
SVCcm = confusion_matrix(y_test, SVCprediction, labels=target_names)
SVCcm_display = ConfusionMatrixDisplay(confusion_matrix = SVCcm, display_labels=target_names)
SVCcm_display.plot(cmap="YlOrBr")
plt.title(f"SVC CM using {max_feature} features")

plt.show()
```

```python
# Confusion matrix
RFcm = confusion_matrix(y_test, RFprediction, labels=target_names)
RFcm_display = ConfusionMatrixDisplay(confusion_matrix = RFcm, display_labels=target_names)
RFcm_display.plot(cmap="RdPu")
plt.title(f"Random Forest CM using {max_feature} features")

plt.show()
```

Detail Result

```python
from sklearn.metrics import classification_report
```

```python
# Decision Tree Result
print(classification_report(y_test, DTprediction, target_names=target_names))
```

```python
# LinearSVC Result
print(classification_report(y_test, SVCprediction, target_names=target_names))
```

```python
# Random Forest Result
print(classification_report(y_test, RFprediction, target_names=target_names))
```

Detail accuracy

```python
print(f'Decision Tree Accuracy: {accuracy_score(y_test, DTprediction)}')
```

```python
print(f'Random Forest Accuracy: {accuracy_score(y_test, RFprediction)}')
```

```python
print(f'Linear SVC Accuracy: {accuracy_score(y_test, SVCprediction)}')
```
